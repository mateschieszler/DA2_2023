# Task 2 - Variance on level of education

### Model 3

Log earnings - education (dummies)

```python
# taking a closer look at distrubution of level of education
data["grade92"].value_counts()
```

```python
# 45 is Professional
# 46 is Phd
# 44 is Masters
```

```python
# Creating dummy variables for three most common categorioes
data["ed_Pro"] = (data["grade92"] == 45).astype(int)
data["ed_Phd"] = (data["grade92"] == 46).astype(int)
data["ed_MA"] = (data["grade92"] == 44).astype(int)
```

```python
reg3 = smf.ols(formula="lnw~female", data=data).fit(cov_type="HC1")
reg4 = smf.ols(formula="lnw~female + ed_Pro + ed_Phd", data=data).fit(cov_type="HC1")
reg5 = smf.ols(formula="lnw~female + ed_Pro + ed_MA", data=data).fit(cov_type="HC1")
```

```python
stargazer = Stargazer([reg3, reg4, reg5])
stargazer.covariate_order(["female", "ed_Pro", "ed_Phd", "ed_MA", "Intercept"])
stargazer.rename_covariates({"Intercept": "Constant"})
stargazer
```

```python
# Since we have data on age, it makes sense to add it to the model. Also add age*female interaction.
# regress separately for men and women
```

```python
reg6 = smf.ols(formula="lnw~age", data=data.query("female==1")).fit(cov_type="HC1")
reg7 = smf.ols(formula="lnw~age", data=data.query("female==0")).fit(cov_type="HC1")
reg8 = smf.ols(formula="lnw~female+age+age*female", data=data).fit(cov_type="HC1")
```

```python
stargazer = Stargazer([reg6, reg7, reg8])
stargazer.covariate_order(["female", "age", "age:female", "Intercept"])
stargazer.rename_covariates({"Intercept": "Constant", "age:female": "female x age"})
stargazer.custom_columns(["Women", "Men", "All"], [1, 1, 1])
stargazer
```

### Model 5 

Log-earnings - education - age groups

```python
reg9 = smf.ols(formula="lnw ~ female + age", data=data).fit(cov_type="HC1")
reg10 = smf.ols(formula="lnw ~ female + ed_Pro + ed_Phd + age", data=data).fit(cov_type="HC1")
reg11 = smf.ols(formula="lnw ~ female + ed_Pro + ed_MA + age", data=data).fit(cov_type="HC1")
```

```python
stargazer = Stargazer([reg9, reg10, reg11])
stargazer.covariate_order(["female","age", "ed_Pro", "ed_Phd", "ed_MA", "Intercept"])
stargazer.rename_covariates({"Intercept": "Constant"})
stargazer
```
